**Import libraries**

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
import chardet
```

**Load Dataset**

```python
with open('/content/ECOMM DATA.csv', 'rb') as f:
    result = chardet.detect(f.read())
    df = pd.read_csv('/content/ECOMM DATA.csv', encoding=result['encoding'])
```

**Convert 'Order Date' and 'Ship Date' to datetime format**

```python
df['Order Date'] = pd.to_datetime(df['Order Date'], format='%d/%m/%Y')
df['Ship Date'] = pd.to_datetime(df['Ship Date'], format='%d/%m/%Y')
```

**sorting values by order date**

```python
df.sort_values(by=['Order Date'],inplace = True, ascending= True)
```

```python
df.set_index('Order Date', inplace = True)
```

```python
print(df.isnull().sum())
```

**To handle null values in postal code. we will not drop them instead we will add the postal code of respective city.**



*   **we need to find the cities for which the postal code is not mentioned**
*   **Fill the postal code of the respective city into the postal column**



```python
df[df['Postal Code'].isnull()]
```

**we can see that the postal code is not mentioned only for Burlington city in vermont state .so we need to fill the postal code of that city.**

```python
df['Postal Code'] = df['Postal Code'].fillna(5401)
```

```python
print(df.isnull().sum())
```

**State-wise Sales**

```python
state_sales = df.groupby(['State'])['Sales'].sum().sort_values(ascending=False)
state_sales.plot(kind='area', figsize=( 10, 5))
plt.title('State-wise Sales')
plt.xlabel('State')
plt.ylabel('Sales')
plt.show()
```

**Top 10 City-wise Sales**

```python
city_sales = df.groupby(['City'])['Sales'].sum().sort_values(ascending=False).head(10)
colors = sns.color_palette("viridis", len(city_sales))
plt.figure(figsize=(12, 5))
plt.scatter(city_sales.index, city_sales.values, color=colors, s=100)  # s is the marker size
plt.title('Top 10 City-wise Sales (Scatter Plot)')
plt.xlabel('City')
plt.ylabel('Sales')
plt.show()
```

**Region-wise Sales**

```python
region_sales = df.groupby(['Region'])['Sales'].sum().sort_values(ascending=False)
colors = sns.color_palette("tab10", len(region_sales))
region_sales.plot(kind='bar', figsize=(10, 5))
colors = ['red', 'green', 'blue','magenta','maroon','cyan','lavender','beige','coral','purple','lime','olive','teal']
region_sales.plot(kind='bar', color=colors, edgecolor='black')
plt.title('Region-wise Sales')
plt.xlabel('Region')
plt.ylabel('Sales')
plt.show()
```

**Sub-Category-wise Sales**

```python
sub_category_sales = df.groupby(['Sub-Category'])['Sales'].sum().sort_values(ascending=False)
colors = sns.color_palette("Set3", len(sub_category_sales))
plt.figure(figsize=(10, 5))
sub_category_sales.plot(kind='bar', color=colors, edgecolor='black')  # edgecolor adds borders to the bars
plt.title('Sub-Category-wise Sales')
plt.xlabel('Sub-Category')
plt.ylabel('Sales')
plt.show()
```

**Segment-wise Sales**

```python
segment_sales = df.groupby(['Segment'])['Sales'].sum().sort_values(ascending=False)
colors = sns.color_palette("Set3", len(segment_sales))
plt.figure(figsize=(6, 8))
plt.pie(segment_sales, labels=segment_sales.index, autopct='%1.1f%%', startangle=140, colors=colors)
plt.title('Segment-wise Sales')
plt.show()
```

**Shipping Modes**

```python
shipping_modes = df['Ship Mode'].value_counts()
shipping_modes.plot(kind='bar', figsize=(8, 5))
colors = ['darkorange', 'yellow', 'green','pink']
shipping_modes.plot(kind='bar', color=colors, edgecolor='black')  # edgecolor adds borders to the bars
plt.title('Shipping Modes')
plt.xlabel('Shipping Mode')
plt.ylabel('Count')
plt.show()
```

**Average shipping time**

```python
df['Order Date'] = pd.to_datetime(df['Order ID'].str[:30], format='%Y%m%d', errors='coerce')
df['Ship Date'] = pd.to_datetime(df['Ship Date'])
df = df.dropna(subset=['Order Date', 'Ship Date'])
df['Shipping Time'] = (df['Ship Date'] - df['Order Date']).dt.days
average_shipping_time = df['Shipping Time'].mean()
print(f"Average Shipping Time: {average_shipping_time} days")
```
